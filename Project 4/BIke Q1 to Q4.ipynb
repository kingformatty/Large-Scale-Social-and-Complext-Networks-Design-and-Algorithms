```python
import numpy as np
import pandas as pd
import seaborn as sns
#from seaborn import stripplot
import matplotlib.pyplot as plt

save_fig = False
```

```python
file = "day.csv"
data_pd = pd.read_csv(file)
pearson_matrix = data_pd.corr(method = 'pearson')
attrs = pearson_matrix.columns

pearson_decimal = round(pearson_matrix, 2)
```

```python
pearson_decimal
```

```python
#Q1
```

```python
def highest_corr(attr, len_attr = 12):
    max_idx = 1
    max_val = abs(attr[1])
    for i in range(1, len_attr):
        if abs(attr[i]) > max_val:
            max_val = abs(attr[i])
            max_idx = i
    return max_idx, max_val
```

```python
#print(attrs)
casual = pearson_matrix['casual'].values
idx_casual, val_casual = highest_corr(casual) 
print("Highest absolute correlation for casual:")
print(attrs[idx_casual], casual[idx_casual])
```

```python
registered = pearson_matrix['registered'].values
idx_registered, val_registered = highest_corr(registered) 
print("Highest absolute correlation for registered:")
print(attrs[idx_registered], registered[idx_registered])
```

```python
cnt = pearson_matrix['cnt'].values
idx_cnt, val_cnt = highest_corr(cnt) 
print("Highest absolute correlation for cnt:")
print(attrs[idx_cnt], cnt[idx_cnt])
```

```python
plt.figure(figsize = (12, 9))
plt.title("Pearson Matrix for bike dataset")
ax = sns.heatmap(pearson_decimal, linewidths=.0, annot=True)
fig = plt.gcf()
plt.show()
if save_fig:
    fig.savefig('Pearson_matrix.png', dpi = 300, bbox_inches = 'tight')
```
