```python
import numpy as np
import pandas as pd
import seaborn as sns
#from seaborn import stripplot
import matplotlib.pyplot as plt

save_fig = True
from sklearn.feature_selection import mutual_info_regression
from sklearn.feature_selection import f_regression
from sklearn.preprocessing import StandardScaler
```

```python
file = "data_transcoding.csv"
data_pd = pd.read_csv(file)

data_pd2 = data_pd.replace({'codec':{'mpeg4':1, 'vp8':2, 'flv':3,'h264':4}, 'o_codec': {'mpeg4':1, 'vp8':2, 'flv':3,'h264':4}})
pearson_matrix = data_pd2.corr(method = 'pearson')
pearson_decimal = round(pearson_matrix, 2)
attrs = pearson_matrix.columns
```

```python
save_fig = False
```

```python
def highest_corr(attr, len_attr = 20):
    max_idx = 0
    max_val = abs(attr[0])
    for i in range(0, len_attr):
        if abs(attr[i]) > max_val:
            max_val = abs(attr[i])
            max_idx = i
    return max_idx, max_val
```

```python
utime_mat = pearson_matrix['utime'].values
#print(utime_mat)
idx_utime, val_utime = highest_corr(utime_mat, 19) 
print("Highest absolute correlation for utime:")
print(attrs[idx_utime], utime_mat[idx_utime])
```

```python
umem_mat = pearson_matrix['umem'].values
idx_umem, val_umem = highest_corr(umem_mat, 19) 
print("Highest absolute correlation for umem:")
print(attrs[idx_umem], umem_mat[idx_umem])
```

```python
plt.figure(figsize = (16, 12))
plt.title("Pearson Matrix for video dataset")
ax = sns.heatmap(pearson_decimal, linewidths=.0, annot=True)
fig = plt.gcf()
plt.show()
if save_fig:
    fig.savefig('Video-Pearson_matrix.png', dpi = 300, bbox_inches = 'tight')
```
